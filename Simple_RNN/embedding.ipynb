```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### SENTENCES
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'Understanding the concept of the words',
]
```

```python
sent
```

```python
## Define the Vocabulary Size ()
voc_size=10000
```

```python
sent
```

```python
## One Hot Encoding Representing ()
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## Word embedding 

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences ## as we know that to train and work onn our model each sentence should have same number of words so to convert different sentences in same number of words we use this library(pad_sequence)
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# Feature Representation
dim=10
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Input

model = Sequential([
    Input(shape=(sent_length,)),     
    Embedding(voc_size, dim)
])

model.compile(optimizer="adam", loss="mse")
model.summary()
```
